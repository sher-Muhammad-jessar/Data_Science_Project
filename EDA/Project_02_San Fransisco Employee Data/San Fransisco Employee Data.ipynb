```python
# droping columns whch are not usefull for this data analysis

data=data.drop(["Id","Status","Notes","Agency"], axis=1)
data.columns
```

```python
data.describe(include='all')
```

```python
# top 6 names whose accurance in data set 

data["EmployeeName"].value_counts().head(6)
```

```python
# find unique job titles 
data["JobTitle"].nunique()
```

```python
# find number of caption in job title 
data[data["JobTitle"].str.contains("CAPTAIN",case=False)].count()
```

```python
# Display all employee names from fire department 
data[data["JobTitle"].str.contains("Fire", case= False)]["EmployeeName"]
```

```python
# fill not provided in NAh feilds
data["EmployeeName"]=data["EmployeeName"].replace("Not provided", np.nan)
data["EmployeeName"]
```

```python
# drop rows having 5 missing values
data.drop(data[data.isnull().sum(axis=1)==5].index, axis=0, inplace=True)
```

```python
# find job title of person with name 
data[data["EmployeeName"]=="ALBERT PARDINI"]["JobTitle"]
```

```python
# benifits of person
data.columns
data[data["EmployeeName"]=="ALBERT PARDINI"]["TotalPayBenefits"]
```

```python
# Display name of person having highest BasePay
# Convert BasePay to numeric (invalid parsing will become NaN)
data["BasePay"] = pd.to_numeric(data["BasePay"], errors='coerce')

data[data["BasePay"] == data["BasePay"].max()]["EmployeeName"]
```

```python
# find average base pay per year 
# Clean BasePay
data["BasePay"] = pd.to_numeric(data["BasePay"], errors='coerce')

# Average BasePay per year
avg_basepay = data.groupby("Year")["BasePay"].mean()
print(avg_basepay)

sb.lineplot(x=avg_basepay.index,y=avg_basepay.values)
plt.title("average base pay graph" )
plt.xlabel("Years")
plt.ylabel("Average Salary")
plt.show()


```

```python
# average base pay pper job title 
data["BasePay"]=pd.to_numeric(data["BasePay"], errors="coerce")
data.groupby("JobTitle")["BasePay"].mean()
```

```python
# find average basepay for accountant 
data["BasePay"]=pd.to_numeric(data["BasePay"], errors="coerce")
data[data["JobTitle"]=="ACCOUNTANT"]["BasePay"].mean()
```

```python
# 5 comman jobs 
ai=data.groupby("JobTitle")["EmployeeName"].count().sort_values(ascending=False).head(5)
sb.barplot(ai,orient="h", color="#FFC0CB")
```

```python
# percentage of employees in top 5  jobs 
plt.pie(ai, autopct="%1.1f%%", labels=ai.index )
plt.title("Top 5 Jobs")
```

```python
plt.hist(data["BasePay"].dropna(), bins=10, edgecolor='000')
plt.title("how base pay spread among all emplyees " )
plt.xlabel("Base Pay")
plt.ylabel("No: of Employees")
plt.show()
```
