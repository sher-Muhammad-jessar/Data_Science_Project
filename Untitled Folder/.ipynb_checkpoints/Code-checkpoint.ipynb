```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the datset 
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# here we are replacing zero with NAN
zero_col=['Pregnancies','Glucose','SkinThickness','Insulin','BMI']
df[zero_col]=df[zero_col].replace(0,np.nan)
```

```python
df.fillna(df.mean(), inplace=True)
df
```

```python
sns.countplot(x='Outcome',data=df)
plt.title("Diabetes Count(0=NO, 1=YES)")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
```

```python
# preparing data for  model training 
x=df.iloc[:,:-1]
y=df['Outcome']
```

```python
y
```

```python
# importing for training model and all 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# now standarizing feature for better performance 
scaler=StandardScaler()
x=scaler.fit_transform(x)

#spiting data in training and testing model 

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# training logistic regression model 
model=LogisticRegression()
model.fit(X_train, y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
print("Classification Report:\n", classification_report(y_test, Y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, Y_pred))
print("Accuracy Score:", accuracy_score(y_test, Y_pred))
```

```python
#now using hyper parameter tuning using gridsearch cv
```

```python
param_grid={
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty':  ['l1','1l2'],
    'solver' :['liblinear']
}

grid_model = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='f1')
grid_model.fit(X_train, y_train)
```

```python
print ("BEST PARAMETERS ",grid_model.best_params_)
```

```python
print("BEST SCORE ",grid_model.best_score_)
```

```python
best_model=grid_model.best_estimator_
```

```python
y_pred_best=best_model.predict(X_test)
```

```python
print("Tuned Model Report:\n", classification_report(y_test, y_pred_best))
```
